```python
# Cell 1 — Imports
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

X_test = pd.read_csv('../data/processed/X_test.csv')
rf = joblib.load('../backend/models/random_forest_model.pkl')

print("✅ Loaded")
```

```python
# Cell 2 — Extract feature importance
importances = rf.feature_importances_
feature_names = X_test.columns.tolist()

importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values('importance', ascending=False)

print(importance_df.head(10))
```

```python
# Cell 3 — Visualize top 15 features
plt.figure(figsize=(10, 6))
sns.barplot(
    data=importance_df.head(15),
    x='importance',
    y='feature',
    palette='coolwarm'
)
plt.title('Top 15 Most Important Features — Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
# Cell 4 — Save importance to file
import os
os.makedirs('../data/processed', exist_ok=True)
importance_df.to_csv('../data/processed/feature_importance.csv', index=False)
print("✅ Saved to data/processed/feature_importance.csv")
```
